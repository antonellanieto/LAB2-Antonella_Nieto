```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Revisión de Cardano desde 2017 hasta fin 2022**

**Creación de DataFrame.**

```python
df = pd.read_csv('cardano_date.csv')
df
```

**Elimino columna que no es útil.**

```python
df = df.drop('Date', axis=1)
df
```

**Verifico si tiene duplicados.**

```python
df.duplicated().sum()
```

**Renombro TimeStamp a Date, ya que previamente transformé la fecha de Timesstamp a formato Date.**

```python
df.rename(columns={'Timestamp': 'Date'}, inplace=True)
df
```

**Verifico si hay campos nulos.**

```python
df.isna().sum()
```

**Precio más alto alcanzado.**

```python
highest = df.loc[df['Price'].idxmax()]
highest
```

**Precio más bajo alcanzado.**

```python
lowest = df.loc[df['Price'].idxmin()]
lowest
```

**Elimino la última fila, ya que corresponde a una fecha que no voy a incluir en el análisis.**

```python
df=df.drop(df.index[-1])
df
```

# **Visualización por año**

# **2017**

```python
X = df['Date']
y = df['Price']

df['Date'] = pd.to_datetime(df['Date'])
df_2017 = df[df['Date'].dt.year == 2017]

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(data=df_2017, x='Date', y='Price')
plt.title('Flujo de precios durante 2017')
plt.xlabel('Fecha')
plt.ylabel('Precio')
plt.xticks(rotation=45)
plt.show()
```
